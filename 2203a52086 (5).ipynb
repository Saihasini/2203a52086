```python
import pandas as pd
import numpy as np
a=pd.read_csv("/content/train.csv")
print(a)
y=a['price_range']
print(y)
z=a.drop('price_range',axis=1)
print(z)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
xtrain_norm=scaler.fit_transform(z)
print(xtrain_norm)
print(xtrain_norm.shape)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xtrain_norm,y,test_size=0.2)
print(xtrain)
print(xtest)
print(ytest)
print(ytrain)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xtrain_norm,y,test_size=0.25)
print(xtrain)
print(xtest)
print(ytrain)
print(ytest)
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```
